#Find total world emission and max country emission for each year

```python
totalWorldemission = df_countries_only.groupby('year').value.sum()

def findpercentageWorld(row):
    row['percentageWorld'] = row.value / totalWorldemission[row.year]
    row['totalWorldEmission'] = totalWorldemission[row.year]
    return row

df = df.apply(findpercentageWorld, axis='columns')

##df["totalWorldEmission"] = df.groupby('year').value.transform('sum')
df["countryMaxEmission"] = df.groupby('country_code').value.transform('max')
df
```

```python
df['percentageWorld'] = df['value'] / df['totalWorldEmission']
df['percentageCountryMax'] = df['value'] / df["countryMaxEmission"]
```

##Redundant

```python
totalWorldemission = df.groupby('year').value.sum()
maxCountryemission = df.groupby('country_code').value.max()

def findpercentageWorld(row):
    row['percentageWorld'] = row.value / totalWorldemission[row.year]
    row['totalWorldEmission'] = totalWorldemission[row.year]
    return row

df = df.apply(findpercentageWorld, axis='columns')

def findpercentageCountryMax(row):
    row['percentageCountryMax'] = row.value / maxCountryemission[row.country_code]
    row['countryMaxEmission'] = maxCountryemission[row.country_code]
    return row

df = df.apply(findpercentageCountryMax, axis='columns')
df
```

```python
import plotly.graph_objects as go
import plotly.express as px
import pandas
from jupyter_dash import JupyterDash
from dash import html, dcc
from dash.dependencies import Input, Output

df = pandas.read_csv('co2_added.csv')


##for year range selection
max_year = df['year'].max()
min_year = df['year'].min()
#print(max_year)
#print(min_year)

##for country dropdown
countries = df['country_name'].unique()
#print(countries)

##for heatmap plot dropdown
heatMapScales = ['WorldContribution', 'CountryMaximum']

##for color options available for heatmap
heatMapColors = [ {'label': c, 'value': c} for c in px.colors.named_colorscales()]


app = JupyterDash(__name__)
app.layout = html.Div([
                ##country-trend
                html.Div([
                    #graph
                    dcc.Graph(id='country-trend',
                              figure = px.line(df, x="year", y="value", color='country_name', markers = True)),
                    ##country selector for trend
                    dcc.Dropdown(countries, [countries[0]], multi=True, id = "country-selector")
                    ]),
                  
                ##world-map  
                html.Div([ 
                    ##map
                    dcc.Graph(id='world-map'),
                    ##colorscale for heatmap
                    html.Label("Colorscale"),
                    dcc.Dropdown(id='colorscale-dropdown', clearable=False, value='plasma', options= heatMapColors),
                    ##heatmapScale
                    html.Label("PercentageBy:"),
                    dcc.Dropdown(heatMapScales, heatMapScales[0], clearable = False, id = "heatmapscale-selector")
                    ]),

                ##Range for years
                #starting year
                html.Div([ 
                    dcc.Input( id="start-year", type='number', placeholder="Start Year"),
                    dcc.Input( id="end-year", type='number', placeholder="End Year")
                    ]),
                ],

                ##style for the main div
                '''style={'display':'flex', 
                       'flex-direction': 'row',
                       'justify-content': 'center',
                       'align-items': 'center'
                       }'''
            )


#callback to update worldmap for colorscale and heatmapscale change
@app.callback(
    Output('world-map', 'figure'),
    Input("colorscale-dropdown", "value"),
    Input("heatmapscale-selector", "value"),
    Input("start-year", "value"),
)
def update_world_map(colorvalue, heatmapScale, start_year):
    df_year = df.loc[df['year'] == start_year]
    if heatmapScale == 'WorldContribution':
      z_val =  df_year['percentageWorld'].astype(float) 
    else:
      z_val = df_year['percentageCountryMax'].astype(float)

    fig  = go.Figure(data=go.Choropleth(
                locations=df_year['country_name'], # Spatial coordinates
                z = z_val, # Data to be color-coded
                locationmode = 'country names', # set of locations match entries in locations
                colorscale = colorvalue
                # colorbar_title = "value",
                )
            )
    return fig


#callback to update country-trend graph
@app.callback(
    Output('country-trend', 'figure'),
    [Input('country-selector', 'value')],)
def update_country_trend(selected_countries):
    print("Hello inside")
    df_country_selected = df.loc[df['country_name'].isin(selected_countries) ]
    ##fig = px.line(df_country_selected, x="year", y="value", color='country_name', markers = False)
    if len(selected_countries) < 10:
      fig = px.line(df_country_selected, x="year", y="value", color='country_name', markers = True,
                    labels=dict(year="Year", value="CO2 emissions", country_name="Country"))
    else:
      fig = px.line(df_country_selected, x="year", y="value", color='country_name', markers = False,
                    labels=dict(year="Year", value="CO2 emissions", country_name="Country"))
    return fig

# Run app and display result inline in the notebook
if __name__== "__main__":
  app.run_server(mode='external', debug=True)
  pass
```
